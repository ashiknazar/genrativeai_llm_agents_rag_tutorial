# üìò Course Syllabus: Generative AI - Foundations and Applications

## üéØ Course Objectives
By the end of this course, students will be able to:
- Understand the fundamental principles behind generative models.
- Analyze and build popular generative models such as VAEs, GANs, and diffusion models.
- Apply generative AI to real-world problems in vision, language, audio, and more.
- Evaluate and mitigate ethical, social, and security implications of generative AI.

---

## üóìÔ∏è Weekly Schedule

### Week 1: Introduction to Generative AI
- What is Generative AI?
- Historical context and applications
- Discriminative vs. Generative Models
- Tools: Python, PyTorch/TensorFlow, Hugging Face

### Week 2: Probability & ML Foundations
- Probability review
- Maximum likelihood estimation
- Bayes Theorem in generative modeling
- Introduction to unsupervised learning

### Week 3: Autoencoders & Variational Autoencoders (VAE)
- Traditional autoencoders
- VAE theory and KL divergence
- Hands-on: Building a VAE

### Week 4: Generative Adversarial Networks (GANs) ‚Äì Fundamentals
- GAN architecture (Generator, Discriminator)
- Training instability and tricks
- Loss functions (minimax, non-saturating)

### Week 5: GAN Variants and Applications
- DCGAN, CGAN, StyleGAN, CycleGAN
- Applications in image synthesis and translation
- Hands-on: Implementing a DCGAN

### Week 6: Diffusion Models
- Overview of diffusion models (DDPM, Stable Diffusion)
- Forward and reverse diffusion
- Applications in image generation

### Week 7: Transformers for Generative Tasks
- Language modeling with GPT
- Autoregressive modeling
- Encoder-decoder vs decoder-only architectures

### Week 8: Large Language Models (LLMs)
- GPT-3, GPT-4, ChatGPT, Claude
- Prompt engineering basics
- Fine-tuning and instruction tuning

### Week 9: Multimodal Generative AI
- Text-to-image (DALL¬∑E, Midjourney)
- Text-to-audio and music generation
- Cross-modal generation challenges

### Week 10: Ethics, Bias, and Safety
- Deepfakes and misinformation
- Copyright and IP issues
- Red-teaming and AI guardrails

### Week 11: Applications and Case Studies
- Use cases in art, healthcare, education, etc.
- Industry examples: OpenAI, DeepMind, Runway

### Week 12: Final Project Presentations
- Project demos and peer feedback
- Wrap-up & discussion of future directions
